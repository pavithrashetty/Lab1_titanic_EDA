##Exploratory Data Analysis(EDA)

```python
#1.Variable PassengerId
df.PassengerId.value_counts()
```

```python
#Finding missing value
df[df.PassengerId.isnull()]
```

```python
#Min,Max,Mean,std of PassengerId
print 'PassengerId'
print 'Min:', df.PassengerId.min()
print 'Max:', df.PassengerId.max()
print 'Mean:', df.PassengerId.mean()
print 'Std:', df.PassengerId.std()
```

<li><B>PassangerId is unique for each Passanger. It is a range of values from 1 to 891 so it is a "continuous" variable.</B>
<li><B>PassangerId has no missing value

```python
#Histogram of the variable PassengerId
df.PassengerId.hist()
```
